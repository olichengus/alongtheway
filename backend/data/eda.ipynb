```python
import pandas as pd
```

```python
stops = pd.read_csv('data/stops.txt', delimiter = ",")
routes = pd.read_csv('data/routes.txt', delimiter = ",")
trips = pd.read_csv('data/trips.txt', delimiter = ",")
stop_times = pd.read_csv('data/stop_times.txt', delimiter = ",")
trips
```

```python
# firstly merge trips with stop_times  
merged_df = pd.merge(trips, stop_times, on='trip_id')  

# then merge the result with stops  
merged_df = pd.merge(merged_df, stops, on='stop_id')  
merged_df
```

```python
# filter the dataframe by route_id  
filtered_df = merged_df[(merged_df['route_id'] == "37810") & (merged_df['direction_id'] == 0)]
filtered_df
```

```python
# get distinct values  
distinct_values = filtered_df[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']].drop_duplicates()
distinct_values
```
